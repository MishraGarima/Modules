## Scraping Wikipedia Page

```python
#importing required packages
import requests
from bs4 import BeautifulSoup
```

```python
#send the get request to the desired url
wiki=requests.get("https://en.wikipedia.org/wiki/World_War_II")

#create a soup object with html parser
soup=BeautifulSoup(wiki.text,'html')

#print data enclosed within title tag
print(soup.find('title'))
```

```python
#obtain a specific division of the page
ww2_contents=soup.find_all("div",class_='toc')
#display contents
ww2_contents
```

```python
#print the text from the division extracted
for i in ww2_contents:
    print(i.text)
```

```python
#extract another division of the page and display the text
ww2_contents=soup.find_all("div",class_='mw-stack desktop-float-right')
for i in ww2_contents:
    print(i.text)
```
