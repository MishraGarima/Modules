# Simple Plots

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

### Create Dummy Dataset

```python
lst = []
for i in range(0,6):
    temp_lst = []
    for i in range(0,9):
        temp_lst.append(np.random.randint(1,100))
    lst.append(temp_lst)  
df1 = pd.DataFrame(lst)

lst = []
for i in range(0,6):
    temp_lst = []
    for i in range(0,9):
        temp_lst.append(np.random.randint(1,100))
    lst.append(temp_lst)    
df2 = pd.DataFrame(lst)

lst1 = []
lst2 = []
lst3 = []
lst4 = []
for i in range(1,11):
    lst4.append(i)
    lst1.append(np.random.randint(10,50))
    lst2.append(np.random.randint(10,30))
    lst3.append(np.random.randint(10,30)) 
df3 = pd.DataFrame({'Column 1':lst4, 'Column 2':lst1, 'Column 3':lst2, 'Column 4':lst3})
```

### Plot - 1

```python
fig,axs=plt.subplots(figsize=(15,5))
sns.heatmap(df1, cmap='magma_r')
sns.heatmap(df1, annot=df1, annot_kws={'va':'bottom','fontsize':12}, fmt="", cbar=False, cmap='magma_r')
sns.heatmap(df1, annot=df2, annot_kws={'va':'top','alpha':0.7}, fmt="", cbar=False,cmap='magma_r')
```

### Plot - 2

```python
fig,axs=plt.subplots(figsize=(15,5))

axs.set_xticks(list(df3['Column 1']))
axs.set_xticklabels(list(df3['Column 1'].astype(str)))
plt.xlabel("Column 1")
plt.ylabel("Column 2")

plt.bar(df3['Column 1'], df3['Column 2'],color='paleturquoise')

a2 = axs.twinx()
a2.plot(df3['Column 1'], df3['Column 3'],'bo--')
a2.plot(df3['Column 1'], df3['Column 4'],'go--')
for i in df3.index:
    plt.text(df3.iloc[i]['Column 1']-0.25, df3.iloc[i]['Column 3']+0.3, df3.iloc[i]['Column 3'], fontsize=15, color='darkgreen')
for i in df3.index:
    plt.text(df3.iloc[i]['Column 1'], df3.iloc[i]['Column 4']+0.3, df3.iloc[i]['Column 4'], fontsize=15, color='darkblue')

plt.legend(["Column 3","Column 4"])
plt.tight_layout()
plt.show()
```

### Plot - 3

```python
fig,axs=plt.subplots(figsize=(15,5))

axs.set_xticks(list(df3['Column 1']))
axs.set_xticklabels(list(df3['Column 1'].astype(str)))
plt.xlabel("Column 2")
plt.ylabel("Column 1")

plt.bar(df3['Column 1'], df3['Column 2'], width=0.45, color='lightsteelblue')
plt.bar(df3['Column 1']+0.45, df3['Column 3'], width=0.45, color='plum')
plt.bar(df3['Column 1']+0.45, df3['Column 4'], width=0.45, bottom=df3['Column 3'],color='pink')

plt.legend(["Column 2","Column 3","Column 4"])
plt.tight_layout()
```
